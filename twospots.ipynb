```python
import numpy as np
import matplotlib.pyplot as plt

from funcs.geometry import create_spherical_grid
from funcs.geometry_spot import get_two_spots




THETA, PHI = create_spherical_grid(int(5000))

lat1, lon1, r1, lat2, lon2, r2 = 0.42, 2.97, 1.36, 1.51, 0.46, 0.42
```

```python
(x, y, z), z_rot, z_rot, amplitude = get_two_spots(THETA, PHI, lat1, lon1, r1, lat2, lon2, r2, np.pi/4)
```

```python

fig = plt.figure(figsize=(10, 5))
spec = fig.add_gridspec(ncols=1, nrows=1)

ax = fig.add_subplot(spec[0, 0], projection='3d')
ax.set_axis_off()

ax.scatter(np.sin(THETA)*np.cos(PHI), np.sin(THETA)*np.sin(PHI), np.cos(THETA), c="grey", s=1,alpha=0.2 )
ax.scatter(x,y,z, c='b', s=1)

# orientation of the image along the x-axis
ax.view_init(0, 0)

ax.quiver(0, 0, 0, z_rot[0], z_rot[1], z_rot[2], color='r')


ax.set_xlim(-2,2)
ax.set_ylim(-2,2)
ax.set_zlim(-1.5,1.5);
```
